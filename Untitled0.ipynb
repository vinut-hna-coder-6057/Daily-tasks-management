```python
def fix_time(x):
    x = str(x).strip()

    if ':' not in x:
        return f"{int(x):02d}:00"

    return x

train['timestamp'] = train['timestamp'].astype(str).apply(fix_time)
test['timestamp'] = test['timestamp'].astype(str).apply(fix_time)

train['timestamp'] = pd.to_datetime(train['timestamp'], format='%H:%M')
test['timestamp'] = pd.to_datetime(test['timestamp'], format='%H:%M')

train['hour'] = train['timestamp'].dt.hour
test['hour'] = test['timestamp'].dt.hour
```

```python
import matplotlib.pyplot as plt

train.groupby('hour')['demand'].mean().plot(marker='o')
plt.xlabel("Hour")
plt.ylabel("Average Demand")
plt.title("Traffic Demand by Hour")
plt.grid(True)
plt.show()
```

```python
import numpy as np



train['minute'] = train['timestamp'].dt.minute
test['minute'] = test['timestamp'].dt.minute

# Time slot (0-95)
train['time_slot'] = train['hour'] * 4 + train['minute'] // 15
test['time_slot'] = test['hour'] * 4 + test['minute'] // 15

# Cyclic encoding
train['hour_sin'] = np.sin(2 * np.pi * train['hour'] / 24)
train['hour_cos'] = np.cos(2 * np.pi * train['hour'] / 24)

test['hour_sin'] = np.sin(2 * np.pi * test['hour'] / 24)
test['hour_cos'] = np.cos(2 * np.pi * test['hour'] / 24)
```

```python
test_ids = test['Index'].copy()
```

```python
train.drop(['Index', 'timestamp'], axis=1, inplace=True)
test.drop(['Index', 'timestamp'], axis=1, inplace=True)
```

```python
train['geo_lane'] = (
    train['geohash'].astype(str)
    + '_'
    + train['NumberofLanes'].astype(str)
)

test['geo_lane'] = (
    test['geohash'].astype(str)
    + '_'
    + test['NumberofLanes'].astype(str)
)
```

```python
train['lane_vehicle'] = (
    train['NumberofLanes'].astype(str)
    + '_'
    + train['LargeVehicles'].astype(str)
)

test['lane_vehicle'] = (
    test['NumberofLanes'].astype(str)
    + '_'
    + test['LargeVehicles'].astype(str)
)
```

```python
train['geo_road'] = (
    train['geohash'].astype(str)
    + '_'
    + train['RoadType'].astype(str)
)

test['geo_road'] = (
    test['geohash'].astype(str)
    + '_'
    + test['RoadType'].astype(str)
)
```

```python
def get_period(h):
    if h < 6:
        return 'night'
    elif h < 12:
        return 'morning'
    elif h < 18:
        return 'afternoon'
    else:
        return 'evening'

train['period'] = train['hour'].apply(get_period)
test['period'] = test['hour'].apply(get_period)
```

```python
from catboost import CatBoostRegressor
X_train = train[train['day'] == 48].drop('demand', axis=1)
y_train = train[train['day'] == 48]['demand']

X_valid = train[train['day'] == 49].drop('demand', axis=1)
y_valid = train[train['day'] == 49]['demand']

cat_features = X_train.select_dtypes(
    include=['object']
).columns.tolist()

print(cat_features)
model = CatBoostRegressor(
    iterations=1000,
    learning_rate=0.05,
    depth=8,
    loss_function='RMSE',
    eval_metric='R2',
    verbose=100
)

model.fit(
    X_train,
    y_train,
    cat_features=cat_features,
    eval_set=(X_valid, y_valid)
)
```

```python
importance = model.get_feature_importance()

imp_df = pd.DataFrame({
    'feature': X_train.columns,
    'importance': importance
}).sort_values('importance', ascending=False)

print(imp_df)
```
